```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from joblib import dump
```

```python
#Se carga el dataset y se muestra
df_data = pd.read_csv("../process_dataset/train_data.csv")
df_data
```

```python
#Se crea una instancia de StandardScaler y se ajusta a los datos
sta_scaler_instance = StandardScaler()
sta_scaler_instance.fit(X=df_data.values)
data_sta_scaled = sta_scaler_instance.transform(X=df_data.values)
```

```python
#Se crea una instancia de RobustScaler y se ajusta a los datos
rob_scaler_instance = RobustScaler()
rob_scaler_instance.fit(X=df_data.values)
data_rob_scaled = rob_scaler_instance.transform(X=df_data.values)
```

```python
#Se crea una instancia de MinMaxScaler y se ajusta a los datos
mm_scaler_instance = MinMaxScaler()
mm_scaler_instance.fit(X=df_data.values)
data_mm_scaled = mm_scaler_instance.transform(X=df_data.values)
```

```python
#Se guardan los datos escalados en un dataframe
df_sta_scaler = pd.DataFrame(data=data_sta_scaled, columns=df_data.columns)
df_rob_scaled = pd.DataFrame(data=data_rob_scaled, columns=df_data.columns)
df_mm_scaled = pd.DataFrame(data=data_mm_scaled, columns=df_data.columns)
```

```python
#Se guardan los dataframe en un archivo csv
df_sta_scaler.to_csv("../process_dataset/train_data_sta_scaled.csv", index=False)
df_rob_scaled.to_csv("../process_dataset/train_data_rob_scaled.csv", index=False)
df_mm_scaled.to_csv("../process_dataset/train_data_mm_scaled.csv", index=False)
```

```python
#Se guardan los objetos scaler en un archivo joblib
dump(sta_scaler_instance, "../results/sta_scaler.joblib")
dump(rob_scaler_instance, "../results/rob_scaler.joblib")
dump(mm_scaler_instance, "../results/mm_scaler.joblib")
```
